```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a planar graph (ensure that it is planar)
G = nx.from_numpy_array(p)

# Calculate the planar layout
pos = nx.planar_layout(G)

# Draw the nodes
nx.draw_networkx_nodes(G, pos, node_size=300, node_color='lightblue')

# Draw the edges
nx.draw_networkx_edges(G, pos, width=1)

# Draw the node labels
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black')

# Show the plot
plt.axis('off')
plt.show()
```

```python
mat_data['tra_X_tr'][0][0].toarray()[1]
df1 = pd.DataFrame(mat_data['tra_X_tr'][0][0].toarray())
for i in range(1,1261):
    df2 = pd.DataFrame(mat_data['tra_X_tr'][0][i].toarray())
    df1=pd.concat([df1, df2], axis=0).reset_index(drop=True)
l=[]
for j in df1.columns:
    l+=[f'C{j}']
df1
```

```python
df3 = pd.DataFrame(mat_data['tra_Y_tr'])
dff=df3.iloc[:,0]
for i in range(1,1261):
    df4=df3.iloc[:,i]
    dff=pd.concat([dff, df4], axis=0).reset_index(drop=True)
l+=["Y1"]
dfp=pd.concat([df1, dff], axis=1)
dfp
```

```python
mat_data['tra_X_tr'][0][0].toarray()[1]
dff1 = pd.DataFrame(mat_data['tra_X_te'][0][0].toarray())
for i in range(1,840):
    dff2 = pd.DataFrame(mat_data['tra_X_te'][0][i].toarray())
    dff1=pd.concat([dff1, dff2], axis=0).reset_index(drop=True)
dff1
```

```python
dff3 = pd.DataFrame(mat_data['tra_Y_te'])
dfff=dff3.iloc[:,0]
for i in range(1,840):
    dff4=dff3.iloc[:,i]
    dfff=pd.concat([dfff, dff4], axis=0).reset_index(drop=True)
dfv=pd.concat([dff1, dfff], axis=1)
df=pd.concat([dfp, dfv], axis=0).reset_index(drop=True)
df.columns=l
dfp.columns=l
dfv.columns=l
df
```

```python
X_train=dfp.iloc[:,:-1]
X_test=dfv.iloc[:,:-1]
y_train=dfp.iloc[:,-1]
y_test=dfv.iloc[:,-1]
dfp
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score
from sklearn.metrics import mean_squared_log_error, median_absolute_error, max_error
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='rbf')
# train the model on the data
svr.fit(X_train, y_train)
# make predictions on the data
y_pred = svr.predict(X_test)

# Calculate regression metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
explained_variance = explained_variance_score(y_test, y_pred)
medae = median_absolute_error(y_test, y_pred)
max_err = max_error(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared (R²):", r_squared)
print("Explained Variance Score:", explained_variance)
print("Median Absolute Error (MedAE):", medae)
print("Max Error:", max_err)
```
